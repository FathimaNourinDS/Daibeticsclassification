```python
df1=pd.get_dummies(df[['fuel','seller_type','transmission','owner','Brand']],dtype=int,drop_first=True)
df1
```

```python
dfe=pd.concat([df,df1],axis=1)
dfe
```

```python
dfe.drop(['name','fuel','seller_type','transmission','owner','Brand'	],axis=1,inplace=True)
dfe
```

```python
dfe.shape
```

```python
dfe.dtypes
```

```python
x=dfe.drop(['selling_price'],axis=1)
x
```

```python
y=dfe['selling_price']
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=1)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
df2=pd.DataFrame({'Actual_value':y_test,'predicted_value':y_pred,'difference':y_test-y_pred})
df2
```

```python
from sklearn.metrics import mean_absolute_percentage_error,r2_score
mape=mean_absolute_percentage_error(y_test,y_pred)
mape
```

```python
r2=r2_score(y_test,y_pred)
r2
```
