```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv('/content/company_sales_data.csv')
df
```

```python
# 1

x=df['month_number']
y=df['total_profit']
plt.plot(x,y)
plt.xticks(x)
plt.yticks([100000,200000,300000,400000,500000])
plt.xlabel('Month')
plt.ylabel('Profit')
plt.title('Monthly-Profit Graph')
```

```python
# 2
x=df['month_number']
y=df['total_profit']
plt.plot(x,y,linestyle='dotted',marker='o',mfc='r',mec='r',color='r',linewidth=3,label='Profit Data')
plt.xlabel('Month')
plt.xticks(x)
plt.ylabel('Solid unit numbers')
plt.title("Month-Graph")
plt.legend(loc='upper right')
```

```python
# 3
month=df['month_number']
facecream=df['facecream']
facewash=df['facewash']
tooth=df['toothpaste']
bath=df['bathingsoap']
shampoo=df['shampoo']
mos=df['moisturizer']
plt.plot(month,facecream,marker='s',color='g',label='Face cream')
plt.plot(month,facewash,marker='s',color='r',label='Face wash')
plt.plot(month,tooth,marker='s',color='y',label='Toothpaste')
plt.plot(month,bath,marker='s',color='k',label='Bathing')
plt.plot(month,shampoo,marker='s',color='m',label='Shampoo')
plt.plot(month,mos,marker='s',color='c',label='Mosturizer')
plt.xticks(month)
plt.legend(loc='upper right')
plt.xlabel('Month')
plt.ylabel('Sales')
```

```python
# 4
x=df['month_number']
y=df['toothpaste']
plt.scatter(x,y)
plt.xticks(x)
plt.xlabel('Month')
plt.ylabel('Toothpaste')
plt.title('Toothpaste Sales')

```

```python
# 5
x=df['month_number']
y=df['facewash']
z=df['facecream']
bar_width=0.35
plt.bar(x,z,width=bar_width,label='Face cream',align='center')
plt.bar(x+bar_width,y,width=bar_width,label='Face wash',align='center')
plt.grid(True)
plt.xticks(df['month_number'])
plt.xlabel('Month')
plt.legend()
```

```python
# 5

plt.figure(figsize=(8,5))
bar_width = 0.35
plt.bar(df['month_number'], df['facecream'], width=bar_width, label='Face Cream Sales', align='center')
plt.bar(df['month_number']+bar_width, df['facewash'], width=bar_width, label='Face Wash Sales', align='center')
plt.xlabel("Month Number")
plt.ylabel("Sales units in number")
plt.title("Face Cream and Face Wash Sales Data")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 6
x=df['total_profit']
plt.hist(x,color='c',edgecolor='black')
plt.xlabel("profit")
plt.ylabel('frequency')
plt.title('Histogram of total profit')
plt.grid()
```

```python
# 6

plt.figure(figsize=(8,5))
plt.hist(df['total_profit'], bins=10, color='skyblue', edgecolor='black')
plt.xlabel("Profit Range")
plt.ylabel("Frequency")
plt.title("Histogram of Total Profit")
plt.grid(True)
plt.show()
```

```python
# x=df['total_units']
# plt.pie(x,labels=x,autopct='%1.01f%%')
# plt.legend()


# totalsales=df.drop(['total_units'],['total_profit'],axis=1).sum()
lab=['facecream','facewash','bath','shampoo','mostizer']
sale=[df['facecream'].sum(),df['facewash'].sum(),df['bathingsoap'].sum(),df['shampoo'].sum(),df['moisturizer'].sum()]
plt.pie(sale,labels=lab,autopct='%1.01f%%')
plt.legend(lab)
plt.title('sales data')
plt.tight_layout()
```

```python
lab = ['Face Cream','Face Wash','Bathing Soap','Shampoo','Moisturizer']
sale = [
    df['facecream'].sum(),
    df['facewash'].sum(),
    df['bathingsoap'].sum(),
    df['shampoo'].sum(),
    df['moisturizer'].sum()
]

plt.figure(figsize=(6,6))
plt.pie(sale, labels=lab, autopct='%1.1f%%', startangle=90)
plt.legend(lab, loc='upper right')
plt.title('Total Sales Distribution (Yearly)')
plt.tight_layout()
plt.show()
```

```python
total_sales = df.drop(['month_number','total_profit','total_units'], axis=1).sum()
plt.figure(figsize=(8,8))
plt.pie(total_sales, labels=total_sales.index, autopct='%1.1f%%', startangle=90)
plt.title("Total Sales Data for Last Year")
plt.show()
```
